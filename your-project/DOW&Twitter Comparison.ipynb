```python
import pandas as pd
import re
```

```python
dow = pd.read_csv("data/DJI-2.csv", sep=";")
twitter = pd.read_csv("data/tweets_clean.csv", sep=";", index_col="Unnamed: 0")
```

```python
dow.head(30)
```

```python
def month_finder(row):
    return re.findall(r"(\d{4}-\d{2})-\d{2}", row["Date"])[0]
dow['Month'] = dow.apply(month_finder, axis=1)
```

```python
dow_peaks = pd.DataFrame(dow.groupby("Month", as_index = False).aggregate({"Low":"min", "High":"max"}))
```

```python
dow_peaks
```

```python
dow_peaks["High_date"] = ""
dow_peaks["Low_date"] = ""
for i in range(len(dow_peaks)):
    low = dow.loc[(dow_peaks["Low"][i]==dow["Low"], "Date")]
    high = dow.loc[(dow_peaks["High"][i]==dow["High"], "Date")]
    dow_peaks["Low_date"][i] = (re.findall(r"\d{4}-\d{2}-\d{2}", str(low)))[0]
    dow_peaks["High_date"][i] = (re.findall(r"\d{4}-\d{2}-\d{2}", str(high)))[0]
```

```python
dow_peaks["high_tweet_links"] = ""
dow_peaks["low_tweet_links"] = ""

for i in range(len(dow_peaks)):
    index_list_low = twitter.loc[dow_peaks["Low_date"][i]==twitter["YYYY-MM-DD"]].index
    links_list_low = []
    index_list_high = twitter.loc[dow_peaks["High_date"][i]==twitter["YYYY-MM-DD"]].index
    links_list_high = []
    if len(index_list_low)>0:
        for j in range(len(index_list_low)):
            links_list_low.append(twitter.loc[dow_peaks["Low_date"][i]==twitter["YYYY-MM-DD"]].loc[index_list_low[j], "link"])
            dow_peaks["low_tweet_links"][i] = links_list_low
    else:
        dow_peaks["low_tweet_links"][i] = "No tweets today!"
    if len(index_list_high)>0:
        for k in range(len(index_list_high)):
            links_list_high.append(twitter.loc[dow_peaks["High_date"][i]==twitter["YYYY-MM-DD"]].loc[index_list_high[k], "link"])
            dow_peaks["high_tweet_links"][i] = links_list_high
    else:
        dow_peaks["high_tweet_links"][i] = "No tweets today!"

        
dow_peaks
```
